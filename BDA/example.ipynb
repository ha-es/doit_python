```python
cheese = pd.DataFrame(
    {
        "first": ["John", "Mary"],
        "last": ["Doe", "Bo"],
        "height": [5.5, 6.0],
        "weight": [130, 150],
    }
)

cheese
```

```python
pd.melt(cheese, id_vars=['first','last'])
```

```python
cheese.melt(id_vars=["first", "last"], var_name="quantity", ignore_index=False)
```

```python
cheese_copy = cheese.melt(id_vars=["first", "last"], var_name="quantity")
cheese_copy
```

```python
cheese_copy.drop_duplicates()
```

```python
cheese_copy.transpose()
```
