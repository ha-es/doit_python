```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) #그래프 4가지를 그리겠다. 
plt.figure(figsize=(13,6))
plt.subplots_adjust(wspace= 0.5, hspace= 0.4) #여러 개 그래프가 그려질 것이니 서로 충돌하지 않게 간격 조절

mpg_features = ['manufacturer','model','drv','category']

#sns 카운트 플랏을 그려보겠다.

for idx, feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    
    sns.countplot(x=feature,
                 data= mpg,
                 hue='drv',
                 palette = 'pastel' , #색상조절하기
                  ax = ax) 
    ax.set_title(f'{feature} Distribution')
```

```python
grid = gridspec.GridSpec(2,2) 
plt.figure(figsize=(18,6))
plt.subplots_adjust(wspace= 0.5, hspace= 0.6)
mpg_features = [ 'year', 'cyl', 'cty', 'hwy']


for idx, feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    
    sns.countplot(x=feature,
                 data= mpg,
                 hue='year',
                 palette = 'bright', 
                  ax = ax,
                  ) 
    ax.set_title(f'{feature} Distribution')
```
