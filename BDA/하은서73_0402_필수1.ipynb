### append()
- 데이터프레임에 행 추가시 사용
- 두 데이터프레임 객체를 행 기준으로 합치는 개념

```python
df = pd.DataFrame([1,2,3,4,5])
df2 = pd.DataFrame([6,7,8,9,10])

print(df)
print(df2)
```

```python
df.append(df2)  #세로(행)으로 결합
```

### describe()
- 다양한 통계량을 요약해준다.

```python
mpg.describe() #수치형 데이터만 요약
```

```python
mpg.describe(include='all') #모든 열의 데이터 출력
```

### drop_duplicates()
- 내용이 중복되는 행을 제거하는 메서드

```python
df.loc[5] = 2  #중복 행 추가
```

```python
df
```

```python
# 1, 5 인덱스의 값이 중복되어 있음
df.drop_duplicates() # 중복행이 제거된 것을 볼 수 있다.
```

### get_values()
- DataFrame 에서 원하는 결과값을 찾을 때 사용

### isin()
-  DataFrame객체의 각 요소가 values값과 일치하는지 여부를 bool형식으로 반환

```python
result = df.isin(df2)  # df와 df2의 values값이 다르므로 결과값이 False
result
```

```python
df3 = pd.DataFrame([0,3,2,4,5])  # df인덱스와 일치는 값 추가
df3
```

```python
result2 = df.isin(df3)  # 3, 4번 인덱스가 True 인 것을 확인할 수 있음
result2
```

### equals()
-  두 객체가 완벽하게 동일한지를 확인하는 메서드

```python
df4 = pd.DataFrame(data=[1,2.0,3,4,5,2])
df4
```

```python
df.equals(df4)  # 데이터 타입이 다르므로 Fasle 반환
```

```python
df4 = pd.DataFrame(data=[1,2,3,4,5,2])
df4
```

```python
df.equals(df4) # df와 타입과 값이 모두 일치하므로 True 반환
```

### sort_values()
-  메서드는 값을 기준으로 레이블을 정렬하는 메서드

```python
df5 = pd.DataFrame(data = [[1,2,np.nan,4,5], [2,3,0,np.nan,8], [6,7,8,9,10]], 
                   columns=['col1', 'col2','col3','col4','col5'], index=['row1','row2','row3'])
df5
```

```python
df5.sort_values(by='col4')  #'by' (col4)기준으로 오름차순 정렬
```

```python
df5.sort_values(by=['col1','col4'])   # col1과 col4 기준으로 오름차순 정렬
```

```python
df5.sort_values(by='row2',axis=1)  # 행 기준 정렬 시 : nan, 문자열과 혼용해서 사용 가능!
```

```python
df5.sort_values(by='col3',ascending=False) # 내림차순
```

```python
df5.sort_values(by='col4',ascending=False, na_position='last') # nan 위치 지정
```

```python
df5.sort_values(by='col4',ignore_index=True)  # index 미사용
```

### to_frame()
-  Series를 DataFrame으로 변환

```python
# Series 이름이 vals인 데이터 생성
df6 = pd.Series(["a", "b", "c"],
              name="vals")
df6
```

```python
# to_frame시 vals이 컬럼의 name이 된다
df6.to_frame()
```
