### 2019~2021년 기온에 따라 제주도를 방문하는 방문객이 연관이 있을까?
##### 가설) 제주도를 겨울보다는 여름에 더 많이 방문할 것이다

```python
# 2019 ~ 2021년 제주시 데이터를 불러옴

df_2019 = pd.read_csv('2019jeju.csv',encoding='euc-kr')
df_2020 = pd.read_csv('2020jeju.csv',encoding='euc-kr')
df_2021 = pd.read_csv('2021jeju.csv',encoding='euc-kr')


print(df_2019)
print(df_2020)
print(df_2021)
```

```python
df_2019.info()
```

```python
# 필요한 칼럼만 가져오기( 2019~ 2021 )
df_2019= df_2019[['일시','평균기온(°C)','평균상대습도(%)','월합강수량(00~24h만)(mm)','평균풍속(m/s)']]
df_2020= df_2020[['일시','평균기온(°C)','평균상대습도(%)','월합강수량(00~24h만)(mm)','평균풍속(m/s)']]
df_2021= df_2021[['일시','평균기온(°C)','평균상대습도(%)','월합강수량(00~24h만)(mm)','평균풍속(m/s)']]
```

```python
# 2019년
vr_2019=df_2019.일시.str.split('-')
vr_2019
```

```python
year = vr_2019.str.get(0)
month = vr_2019.str.get(1)
```

```python
df_2019['year']= year
df_2019['month'] = month

df_2019
```

```python
# 2020년
vr_2020=df_2020.일시.str.split('-')

year = vr_2020.str.get(0)
month = vr_2020.str.get(1)

df_2020['year']= year
df_2020['month'] = month

```

```python
df_2020
```

```python
# 2021년
vr_2021=df_2021.일시.str.split('-')

year = vr_2021.str.get(0)
month = vr_2021.str.get(1)

df_2021['year']= year
df_2021['month'] = month
```

```python
df_2021
```

```python
df=pd.concat([df_2019, df_2020, df_2021])
df
```

```python
df.groupby('month').mean()
```

```python
df.groupby('year').mean()
```
