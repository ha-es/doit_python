### 2019~2021년 제주시 데이터 / 제주시 교통사고 데이터
##### 가설) 제주시에 발생하는 교통사고는 '여름'이 많을 것 이다.

```python
# 2019 ~ 2021년 제주시 데이터를 불러옴

df_2019 = pd.read_csv('2019jeju.csv',encoding='euc-kr')
df_2020 = pd.read_csv('2020jeju.csv',encoding='euc-kr')
df_2021 = pd.read_csv('2021jeju.csv',encoding='euc-kr')

print(df_2019)
print(df_2020)
print(df_2021)
```

```python
df_2019.info()
```

```python
# 필요한 칼럼만 가져오기( 2019~ 2021 )
df_2019= df_2019[['일시','평균기온(°C)','평균상대습도(%)','월합강수량(00~24h만)(mm)','평균풍속(m/s)']]
df_2020= df_2020[['일시','평균기온(°C)','평균상대습도(%)','월합강수량(00~24h만)(mm)','평균풍속(m/s)']]
df_2021= df_2021[['일시','평균기온(°C)','평균상대습도(%)','월합강수량(00~24h만)(mm)','평균풍속(m/s)']]
```

```python
# 2019년
vr_2019=df_2019.일시.str.split('-')
vr_2019
```

```python
year = vr_2019.str.get(0)
month = vr_2019.str.get(1)
```

```python
df_2019['year']= year
df_2019['month'] = month

df_2019
```

```python
# 2020년
vr_2020=df_2020.일시.str.split('-')

year = vr_2020.str.get(0)
month = vr_2020.str.get(1)

df_2020['year']= year
df_2020['month'] = month

```

```python
df_2020
```

```python
# 2021년
vr_2021=df_2021.일시.str.split('-')

year = vr_2021.str.get(0)
month = vr_2021.str.get(1)

df_2021['year']= year
df_2021['month'] = month
```

```python
df_2021
```

```python
# 데이터 합치기
df=pd.concat([df_2019, df_2020, df_2021])
df
```

```python
# 월 별 평균
df_mean = df.groupby('month').mean()
df_mean
```

### 제주시 교통사고 통계

```python
jeju_ac = pd.read_csv('시도별_교통사고.csv',encoding='euc-kr')
jeju_ac
```

```python
# 시도별 데이터이므로 제주를 확인하기 위해 tail
jeju_ac.tail(20)
```

```python
# '제주'만 출력하기
jeju_ac = jeju_ac.loc[209:221]
jeju_ac
```

```python
# 불필요한 행 삭제
jeju_ac.drop(209, inplace=True)
```

```python
jeju_ac.drop('시도별(1)', axis=1, inplace=True)
```

```python
# 연도가 'object' 형식으로 되어 있다는 것을 확인할 수 있음
jeju_ac.info()
```

```python
# int형으로 형변환
jeju_ac[['2019','2019.1','2019.2','2020','2020.1','2020.2','2021','2021.1','2021.2']]=\
    jeju_ac[['2019','2019.1','2019.2','2020','2020.1','2020.2','2021','2021.1','2021.2']].apply(pd.to_numeric)
```

```python
jeju_ac.info()
```

```python
# 월 별 합계를 구함
jeju_ac['총 합계'] = jeju_ac.sum(axis=1)
jeju_ac
```

```python
# 컬럼 이름 바꾸기
jeju_ac.rename(columns = {'월별(1)' : 'month'}, inplace = True)
jeju_ac
```

```python
# 필요한 행만 추출
jeju_ac = jeju_ac[['month','총 합계']]
jeju_ac
```

```python
# month를 인덱스로 설정함
jeju_ac = jeju_ac.set_index('month')
```

```python
jeju_ac.transpose()
```

```python
pd.pivot_table(jeju_ac, columns='month', values='총 합계')
```

```python
jeju_ac.sort_values(by='총 합계', ascending=False)
```
