- 범례 값 만들어서 비교해 보기!
- left
- center
- right
- upper left , right, center
- lower left, right, center

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.scatter([2,4,6],[1,3,5], label='A')
plt.scatter([3,5,7],[3,3,3], label='B')
plt.scatter([3,2,5],[1,2,3], label='C')
plt.legend()
```

```python
plt.scatter([2,4,6],[1,3,5], label='A')
plt.scatter([3,5,7],[3,3,3], label='B')
plt.scatter([3,2,5],[1,2,3], label='C')
plt.legend(loc='center')
```

```python
plt.scatter([2,4,6],[1,3,5], label='A')
plt.scatter([3,5,7],[3,3,3], label='B')
plt.scatter([3,2,5],[1,2,3], label='C')
plt.legend(loc='best')  # 'best'가 default값
```

```python
plt.scatter([2,4,6],[1,3,5], label='A')
plt.scatter([3,5,7],[3,3,3], label='B')
plt.scatter([3,2,5],[1,2,3], label='C')
plt.legend(loc='right')
```

```python
plt.scatter([2,4,6],[1,3,5], label='A')
plt.scatter([3,5,7],[3,3,3], label='B')
plt.scatter([3,2,5],[1,2,3], label='C')
plt.legend(loc='upper left')
```

```python
plt.scatter([2,4,6],[1,3,5], label='A')
plt.scatter([3,5,7],[3,3,3], label='B')
plt.scatter([3,2,5],[1,2,3], label='C')
plt.legend(loc='upper right')
```

```python
plt.scatter([2,4,6],[1,3,5], label='A')
plt.scatter([3,5,7],[3,3,3], label='B')
plt.scatter([3,2,5],[1,2,3], label='C')
plt.legend(loc='upper center')
```

```python
plt.scatter([2,4,6],[1,3,5], label='A')
plt.scatter([3,5,7],[3,3,3], label='B')
plt.scatter([3,2,5],[1,2,3], label='C')
plt.legend(loc='lower left')
```
