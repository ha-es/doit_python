```python
# merge -> 데이터 병합
a = pd.DataFrame({'name': ['김철수','김영희','홍길동'],
                 '영어 점수' : [10,20,30]})

b = pd.DataFrame({'name': ['김철수','김영희','홍길길'],
                 '수학 점수' : [10,20,40]})
```

```python
display(a)
display(b)
```

```python
# merge사용 시 -> 공통된 기준 필요
total = pd.merge(a, b, how='inner') #name 공통된 키라고 이야기함 (pk)
total
```

```python
# concat
pd.concat([a,b]) #디폴트값 -> 0
```
