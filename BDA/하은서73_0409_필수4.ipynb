```python
wine.info()
```

1. alcohol -> 알코올 도수
2. malic_acid -> 사과산이라 불리며 좋은 향이 남
3. ash -> '재'를 의미하며 와인의 보관 과정에서 재의 양에 따라 향과 맛에 영향을 줌
4. alcalinity_of_ash -> 알칼리성 재, 포도에서 나온 부산물 중 하나로, 와인의 PH와 관계가 있음
5. magnesium -> 마그네슘
6. total_phenols -> 와인의 색상과 맛에 영향을 미치는 중요한 화합물인 페놀의 총량
7. flavanoids	-> 플라보노이드로 포도의 껍질과 씨앗에서 추출됨, 와인의 맛과 향을 개선시키는데 기여
8. nonflavanoid_phenols -> 비-폴라보노이드 페놀화합물로, 주로 포도 껍질에서 추출되며, 와인의 맛,향,색상에 영향을 끼침
9. proanthocyanins -> 프로안토시아닌으로 와인의 맛과 향, 색상을 결정함, 농도가 진함과 색상이 비례함
10. color_intensity -> 색깔의 강렬함
11. hue -> 와인의 색조를 나타내는 지표
12. od280/od315_of_diluted_wines -> 희석된 와인의 흡광비로 와인의 색상과 와인 안의 단백질 농도를 추정하는데 사용됨
13. Proline -> 와인 안에 존재하는 아미노산 중 하나로 산화에 대한 내성을 높이는데 영향을 끼침

### NAN처리

```python
# class가 0인 와인의 alcohol의값이 null, 따라서 평균값으로 대체
wine = ( wine.assign(alcohol = np.where(wine['alcohol'].isna(), 13.7352, wine['alcohol'])))
wine
```

```python
wine.isna().sum()
```

```python
wine[wine['nonflavanoid_phenols'].isna()]
```

```python
wine.rename(columns = {'class' : 'class_'}, inplace = True)
```

```python
wine = ( wine.assign(nonflavanoid_phenols = np.where(wine['nonflavanoid_phenols'].isna(), 0.365, wine['nonflavanoid_phenols'])))
wine.isna().sum()
```

```python
# hue -> 와인의 발효와 숙성정도를 판단하는 지표
# hue의 값이 0.5 이하 -> 미숙한 와인
# 1.0 이상 : 와인이 산화되었다고 판단
```

```python
# 0.5 이하 : poor
# 1.0 이상 : oxidize
# 아니면 good

wine['hue_test'] = np.where( wine['hue'] <= 0.5 , 'poor',
                           (np.where( wine['hue'] > 1.0 , 'oxidize', 'good')) )
wine.head()
```

```python
wine
```

```python
plt.figure(figsize=(10,5))
q = sns.countplot(data=wine, x='hue_test', hue='class')
```

```python
sns.heatmap(wine.corr(), annot = True)
```

```python
import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) 
plt.figure(figsize=(30,20))
plt.subplots_adjust(wspace= 0.5, hspace= 0.4) 

wine_ = ['alcohol','ash','alcalinity_of_ash','proanthocyanins']

for idx, feature in enumerate(wine_):
    ax = plt.subplot(grid[idx])
    
    sns.histplot(x=feature,
                 data= wine,
                 hue='class_',
                 palette = 'pastel' ,
                  kde = True,
                  ax = ax) 
    ax.set_title(f'{feature} visualization')
```

```python
import scipy.stats as stats

fig = plt.figure(figsize = (20,8))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

stats.probplot(wine.query('class_ == 0')['alcohol'], plot=ax1)
stats.probplot(wine.query('class_ == 1')['alcohol'], plot=ax2)
stats.probplot(wine.query('class_ == 2')['alcohol'], plot=ax3)
plt.show()
```

```python
import matplotlib as mpl
import seaborn.objects as so

f = mpl.figure.Figure(figsize=(10, 5))
sf1,sf2 = f.subfigures(1,2)
(
    so.Plot(wine, x = 'alcohol', color = 'class_')
    .add(so.Line(),so.Hist(stat = 'proportion',binwidth = 0.25,common_norm = False))
    .scale(color="flare")
    .on(sf1)
    .plot()    
)
(
    so.Plot(wine, x='alcohol')
    .facet("class_")
    .add(so.Bars(), so.Hist(stat = 'proportion',binwidth = 0.25,common_norm = False))
    .on(sf2)
    .plot()
)
```
