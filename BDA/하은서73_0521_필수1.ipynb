- 시계열데이터 문법
- offsets, first 근로기준의 날 뽑을 수도 있고, 첫주의 월요일만 뽑을 수 있고 
- 리서치 하셔서 정리하시고 최소 5개 이상 문법 찾아서 코드와 주석처리 해주시면 됩니다.

```python
import pandas as pd
import datetime
```

```python
crime = pd.read_csv('crime.csv')
crime
```

```python
crime['REPORTED_DATE'] = pd.to_datetime(crime['REPORTED_DATE'])
```

```python
crime.info()
```

```python
# 시계열 데이터 인덱스로 저장
crime_dt =crime.set_index('REPORTED_DATE')
```

```python
crime_dt15=crime_dt.loc['2015'].sort_index()
crime_dt15
```

- 2015년 1월 1일 기준으로 7일동안

```python
crime_dt15.first('7D')
```

- 2015년 마지막 날짜 기준으로 7일

```python
crime_dt15.last('7D')
```

- 2015년 기준으로 날짜를 5일씩 이동하여 데이터 출력

```python
crime_dt15.shift(freq='5D')
```

- 2015년 '수요일'만 추출
- 월요일:0, 화요일:1, 수요일:2 ...

```python
w = crime_dt15[crime_dt15.index.weekday==2]
w
```

- 2015년 '수요일' 중에서 '오전'에 일어난 사건 추출

```python
w[(w.index.hour >=0) & (w.index.hour <12)]
```
