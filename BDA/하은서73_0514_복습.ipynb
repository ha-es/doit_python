#### comcat 함수
- df + df 합치는 경우
- axis 축, igonre_index 병합 후 index 새롭게 만들지 유/무

```python
import pandas as pd

df1 = pd.DataFrame({'A':[1,2,3,4], 'B':[1,2,3,4]})
df2 = pd.DataFrame({'A':[3,4], 'B':[3,4]})
df3 = pd.DataFrame({'A':[2,6], 'B':[5,3]})


display(df1)
display(df2)
display(df3)

```

```python
pd.concat([df1,df2,df3], ignore_index=True)
```

```python
pd.concat([df1,df2,df3], axis=1)
```

- concat은 컬럼이 동일한 경우 쉽게 붙일 수 있음
- 시계열 데이터 유의

### merge 함수
- 두 개의 데이터 프레임 join하는 함수
- 공통된 컬럼을 기준으로 df을 병합 가능

```python
df1 = pd.DataFrame({'이름' : ['홍길동','김영희','김철수','임꺽정'],
                    '영어' : [100,40,70,60]})

df2 = pd.DataFrame({'이름' : ['홍길동','김영희','김철수'],
                    '수학' : [50,100,70]})
```

```python
df1
```

```python
df2
```

```python
# inner 교집합

pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, on='이름')
```

```python
df3 = pd.DataFrame({'이름' : ['홍길동','김영희','김철수','임꺽정'],
                    '수학' : [50,100,70,60],
                    '생년월일' : ['00','02','01','03']})

df4 = pd.DataFrame({'이름' : ['홍길동','김영희','김철수'],
                    '수학' : [50,100,70],
                    '지각횟수' : [0,1,2]})
```

```python
df3
```

```python
df4
```

```python
# on 사용시 2개이상 컬럼 사용 가능

pd.merge(df3,df4, on=['이름','수학'])
```

```python
pd.merge(df3,df4, on='이름')
```

```python
df3 = pd.DataFrame({'이름' : ['홍길동','김영희','김철수','임꺽정'],
                    '수학' : [50,100,70,60],
                    '생년월일' : ['00','02','01','03']})

df4 = pd.DataFrame({'name' : ['홍길동','김영희','김철수'],
                    '수학' : [50,100,70],
                    '지각횟수' : [0,1,2]})
```

```python
# right_on, left_on

pd.merge(df3, df4, left_on=['이름'], right_on=['name'])
```

```python
pd.merge(df3, df4, left_on=['이름', '수학'], right_on=['name','수학'])
```

```python
# left_index, right_index 

df3_sp = df3.set_index('이름')
```

```python
df3_sp.index
```

```python
df3_sp
```

```python
df4
```

```python
pd.merge(df3_sp, df4, left_index=True, right_on='name')
```

### How
- join 키에 대한 inner 기본값
- left
- right
- outer : 왼쪽과 오른쪽 df 조인키에 대한 합집합 기준

```python
df1 = pd.DataFrame({'A':[1,2,3],'B':[1,2,3]})
df2 = pd.DataFrame({'A':[2,3,4],'C':[7,8,9]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, on='A', how='inner')
```

```python
pd.merge(df1,df2, on='A')
```

```python
pd.merge(df1,df2, on='A', how='left')
```

```python
pd.merge(df1,df2, on='A', how='right')
```

```python
pd.merge(df1,df2, on='A', how='outer')
```
