```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
```

```python
sns.set(style = 'dark')
plt.plot([1,2,3],[4,5,6],label='mpg.cty', color='c',linestyle='dashed', linewidth=1, marker='o')
plt.plot([6,7,8],[3,2,5],label='mpg.hwy',color='r',linestyle='dotted', linewidth=1, marker='>')
plt.legend()
```

```python
sns.set(style = 'darkgrid')
plt.plot([1,2,3],[4,5,6],label='mpg.cty', color='g',linestyle='-.', linewidth=1, marker='s')
plt.plot([6,7,8],[3,2,5],label='mpg.hwy',color='b', linewidth=2, marker='h')
plt.legend()
```

```python
x = np.random.normal(0,1,5000)
y = np.random.normal(0,1,5000)

x1 = np.random.normal(0,1,1000)
y1 = np.random.normal(0,1,1000)
```

```python
plt.scatter(x,y, alpha= 0.05)
```

```python
plt.scatter(x, y, alpha=0.1, color='b',edgecolors='b')
plt.scatter(x1, y1, alpha=0.1, color='r',edgecolors='r')
```

```python
plt.scatter(x, y, alpha=0.1, color='blue',edgecolors='black',linewidths=2)
plt.scatter(x1, y1, alpha=0.1, color='blue',edgecolors='red',linewidths=5)
```
