### 다른 방법으로 누락값 대체하는 법

- fillna(method = ' ' ) 문법 공식
- ffill -> 누락값이 나타나기 전의 값으로 누락값을 변경
- bfill -> 누락값이 나타난 이후 첫 번째 값으로 앞쪽의 누락값 변경

```python
ebola.fillna(method='ffill').iloc[0:10, 0:5]['Cases_Liberia']
```

```python
ebola.fillna(method='bfill').iloc[0:10, 0:5]['Cases_Liberia']
```

```python
ebola.iloc[0:10, 0:5]['Cases_Liberia'].mean()
```

```python
# 보간법
ebola.interpolate().iloc[0:10, 0:5]['Cases_Liberia']
```

```python
# 양쪽에 있는 값을 이용해 중간값을 구한 후 누락값 처리
ebola.interpolate(method='quadratic').iloc[0:10, 0:5]['Cases_Liberia'].mean()
```

### 누락값 대체 법

- 보간법
- interpolate(method=' ')
- 선형보간, spline 보간법 등등

```python
ebola['Cases_Liberia'].plot()
```

```python
# 평균으로 대체 시
ebola.fillna(8049.2)['Cases_Liberia'].plot()
```

```python
# 스플라인 보간법
# 각 데이터의 포인트 구간별로 낮은 숮ㄴ의 다항식 보간을 사용함
# 직선으로 연결되는 것 보다 좀 더 부드럽게 연결
ebola.interpolate(method='quadratic')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method='dsfd')['Cases_Liberia'].plot()
```

### melt 함수

- melt 메서드 인자
- id_vars : 위치를 그대로 유지할 열의 이름을 지정
- value_vars : 행으로 위치를 변경할 열의 이름을 지정
- var_name :value_vars로 위치 변경할 열의 이름
- value_name : var_ name의 위치 변경할 열의 데이터를 저장한 열의 이름

```python
pew = pd.read_csv('pew.csv')
pew
```

```python
# id_vars : 축
pd.melt(pew, id_vars='religion')
```

```python

pd.melt(pew, id_vars='religion', var_name='inconme', value_name='count')
```

### 해당 컬럼을 유의미하게 변수를 만들 수 있음

- 컬럼에 있는 의미를 분리해 새로운 변수를 만들어 사용할 수 있음
- 기존에 있는 변수를 split 나눠서 새로운 변수 두 개로 만들기
- 피처엔지니어링 개념으로 이해해서 머신에게 학습시킬 때 좀 더 잘 학습시키기 위해서 나누는 과정

```python
ebola_sp = pd.melt(ebola, id_vars=['Date','Day'])
ebola_sp
```

```python
vr_sp = ebola_sp.variable.str.split('_')
vr_sp
```

```python
# str 문법 중 get( )
case = vr_sp.str.get(0)
case
```

```python
case_cr = vr_sp.str.get(1)
case_cr
```

```python
ebola_sp['case']= case
ebola_sp['case_cr'] = case_cr
```

```python
ebola_sp
```
