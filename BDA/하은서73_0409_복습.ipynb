### 히트맵

- 2차원 배열로 각 요소의 색상을 시각화하는 것!

```python
data = pd.DataFrame({'DA' : [50,20,55,88], 'PG' : [11,22,33,44]},
                     index = ['A','B','C','D'])

data
```

```python
sns.heatmap(data)
```

```python
# 'd' 정수 표시, annot 박스 대응 데이터 표시
sns.heatmap(data, annot=True, fmt='d')
```

```python
sns.heatmap(data, annot=True, fmt='d', linewidth=3, linecolor='w')
```

```python
plt.figure(figsize=(4,5))
sns.heatmap(data, annot=True, fmt='d', linewidth=3,linecolor='w', cmap='Greys')
```

### 다양한 시각화 그래프 여러 개 그리기

- subplots 통해서 그래프 여러 개 그리기
- fig : 도화지 개념
- axis, ax : 실제 그림 그리기

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8,4))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax[0].plot(mpg['cty'])
ax[1].hist(mpg['cty'].value_counts(), facecolor='r')
```

### 반복문을 통한 다양한 그래프 그리기

```python
import matplotlib.gridspec as gridepec
grid = gridepec.GridSpec(2,2)
plt.figure(figsize=(15,10))
plt.subplots_adjust(wspace=0.5, hspace=0.4)

mpg_features = ['manufacturer', 'model','category','drv']



for idx, feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    
    sns.countplot( x=feature,
                  data=mpg,
                  hue = 'drv',
                  palette= 'pastel',
                  ax = ax)
    
    ax.set_title(f'{feature} Distribution')
    
    
```
