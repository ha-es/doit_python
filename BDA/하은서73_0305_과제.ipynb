### Q1. 'audi'의 구동방식(drv)별 도시연비(cty)의 평균은?

```python
# mpg_audi 변수를 만든 후 manufacturer,drv집단으로 묶어 cty의 평균을 구한다.
mpg_audi = mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty','mean'))
mpg_audi
```

```python
# 그 중 'audi'의 값을 가져와야 하므로 query문을 사용
mpg_audi.query('manufacturer =="audi"')
```

```python
# 결론 : 'audi' 사륜구동(4)의 평균은 16.82, 전륜구동(f)의 평균은 18.86이다.
# 따라서 전륜구동의 도시연비 평균이 더 높다.
```

### Q2. 'audi'와 'toyota' 중 어느 제조회사 (manufacturer)의 도시연비(cty)의 평균이 더 높을까?

```python
# 'mpg_audi'변수 생성 후 query문으로 audi 데이터만 가져온다
mpg_audi = mpg.query('manufacturer == "audi"')
mpg_audi
```

```python
# 'mpg_toyota'변수 생성 후 query문으로 toyota 데이터만 가져온다
mpg_toyota = mpg.query('manufacturer == "toyota"')
mpg_toyota
```

```python
# audi의 cty 평균
mpg_audi['cty'].mean()
```

```python
# toyot의 cty 평균
mpg_toyota['cty'].mean()
```

```python
# 결론 : toyota의 평균 도시연비가 더 높다.
```

### Q3. 자동차 모델(model)별 평균 배기량(displ)이 가장 높은 모델과 가장 낮은 모델 3곳은?

```python
# 'mpg_model'변수 생성 후 dispal의 평균을 구한다.
# groupby()의 as_index = False는 변수를 인덱스로 바꾸지 않고 원래대로 유지
# mean_displ(평균 배기량)의 값을 내림차순으로 정렬
mpg_model = mpg.groupby('model', as_index=False).agg(maen_displ = ('displ','mean')).sort_values('maen_displ', ascending=False)
mpg_model
```

```python
# mean_displ 가장 높은 곳 (상위 3곳 )
mpg_model.head(3)
```

```python
# mean_displ가장 낮은 곳 (하위 3곳 )
mpg_model.tail(3)
```

```python
# 결론 : 평균 배기량이 가장 높은 곳은 corvette -> k1500 tahoe 4wd -> c1500 suburban 2wd 이며,
# 가장 낮은 곳은 'civic' 이다.
```

### Q4. 어떤 제조회사 (manufacturer)에서 'suv'차종을 를 가장 많이 생산할까?

```python
# 'mpg_suv'변수에 'suv'추출 후 제조회사(manufacturer)별로 개수 출력
mpg_suv = mpg.query('category=="suv"').value_counts('manufacturer')
mpg_suv
```

```python
# 결론 : chevrolet와 ford가 9개로 가장 많이 suv를 생산한다.
```

### Q5. 자동차 종류(category)별 평균 고속도로 연비(hwy)가 가장 높은 곳은?

```python
# 'mpg_category' 변수에 category별로 묶은 후, 평균 고속도로 연비를 구한다.
# 가장 높은 곳을 추출하기 위해 내림차순 정렬
mpg_category = mpg.groupby('category').agg(mean_hwy = ('hwy','mean')).sort_values('mean_hwy', ascending=False)
mpg_category.head(1)
```

```python
# 결론 : compact가 평균 고속도로 연비(mean_hwy)가 가장 높다.
```
