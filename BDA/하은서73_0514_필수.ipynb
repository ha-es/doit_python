### jeju_income 데이터

```python
# 내국인 방문객과 교통량에 대한 관계
# 방문객이 많으면 교통량이 증가할 것이다 ?

jeju_income = pd.read_csv('제주_내국인관광객현황_2021.csv', sep=',',encoding='euc-kr')
jeju_income
```

```python
jeju_income['sum'] = jeju_income['행태별(개별여행)'] + jeju_income['행태별(부분패키지)'] + jeju_income['행태별(패키지)']
jeju_income
```

```python
jeju_income.info()
```

```python
jeju_income['데이터기준일자'] = jeju_income['데이터기준일자'].astype('str')
```

```python
jeju_income['데이터기준일자']=pd.to_datetime(jeju_income['데이터기준일자'])
```

```python
jeju_income.info()
```

```python
# 년 만 가져오기

jeju_income['year'] = jeju_income['데이터기준일자'].dt.year
jeju_income.head()
```

```python
# 필요한 데이터만 추출

jeju_income = jeju_income[['구분연월', 'sum', 'year']]
jeju_income
```

```python
jeju_income = jeju_income.rename(columns= { '구분연월' : 'month'})
jeju_income
```

```python
jeju_income.loc[0,'month'] =1
jeju_income.loc[1,'month'] =2
jeju_income.loc[2,'month'] =3
jeju_income.loc[3,'month'] =4
jeju_income.loc[4,'month'] =5
jeju_income.loc[5,'month'] =6
jeju_income.loc[6,'month'] =7
jeju_income.loc[7,'month'] =8
jeju_income.loc[8,'month'] =9
jeju_income.loc[9,'month'] =10
jeju_income.loc[10,'month'] =11
jeju_income.loc[11,'month'] =12


jeju_income
```

```python
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'


plt.figure(figsize = (10, 5))
sns.barplot(data = jeju_income ,x = 'month', y = 'sum')

plt.title('2021년 월 별 방문객 수')
plt.xticks(rotation= 0)
plt.show
```

```python
df_mg = pd.merge(left=df_2021,right=jeju_income,on="month",how="left")
df_mg
```
