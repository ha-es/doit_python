### 2021년 방문객

```python
# 내국인 방문객과 교통량에 대한 관계
# 방문객이 많으면 교통량이 증가할 것이다 ?

jeju_2021 = pd.read_csv('제주_내국인관광객현황_2021.csv', sep=',',encoding='euc-kr')
jeju_2021
```

```python
jeju_2021['sum'] = jeju_2021['행태별(개별여행)'] + jeju_2021['행태별(부분패키지)'] + jeju_2021['행태별(패키지)']
jeju_2021
```

```python
jeju_2021.info()
```

```python
# 필요한 데이터만 추출

jeju_2021 = jeju_2021[['구분연월', 'sum']]
jeju_2021
```

```python
jeju_2021 = jeju_2021.rename(columns= { '구분연월' : 'month'})
jeju_2021
```

```python
for i in jeju_2021.index:
    jeju_2021.loc[i,'month']= i+1
    

jeju_2021
```

```python
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'


plt.figure(figsize = (10, 5))
sns.barplot(data = jeju_2021,x = 'month', y = 'sum')

plt.title('2021년 월 별 방문객 수')
plt.xticks(rotation= 0)
plt.show
```

```python
df_mg_group_2021 = df_2021['target'].groupby(df_2021['month'])
mean = df_mg_group_2021.mean()
mean
```

```python
df_mg2 = pd.merge(left=mean,right=jeju_2021,on="month",how="left")
df_mg2
```

```python
df_mg2.info()
```

```python

# 그래프 그리기
fig, ax1 = plt.subplots(figsize = (10, 7))


# 막대 그래프 그리기
ax1.bar(df_mg2['month'], df_mg2['sum'], color='blue')
ax1.set_xlabel('month')
ax1.set_ylabel('sum', color='blue')
ax1.tick_params('y', colors='blue')
ax1.set_xticks(df_mg2['month'])


# 선 그래프 그리기
ax2 = ax1.twinx()
ax2.plot(df_mg2['month'], df_mg2['target'], color='coral')
ax2.set_ylabel('target', color='red')
ax2.tick_params('y', colors='red')

# 그래프 출력
plt.show()
```
