```python
# 데이터 불러오기
import pandas as pd

df = pd.read_csv('news_comment_BTS.csv', encoding='UTF-8')
df.info()
```

```python
# 불필요한 문자 제거
import re
df['reply'] = df['reply'].str.replace('[^가-힣]',' ', regex=True)
df['reply'].head()
```

```python
# 명사 추출
# kkma(꼬꼬마) : 형태소 분석기는 띄어쓰기 오류가 있는 문장에서도 형태소를 잘 추출하는 장점

import konlpy
kkma = konlpy.tag.Kkma()
```

```python
# 명사 추출 -  apply()사용
nouns = df['reply'].apply(kkma.nouns)
nouns
```

```python
# 단어 빈도표 만들기
nouns = nouns.explode()
nouns
```

```python
# 데이터 프레임 만들기
df_word = pd.DataFrame( { 'word' : nouns})

# 글자 수 추가
df_word['count'] = df_word['word'].str.len()

# 두 글자 이상 단어만 남겨두기
df_word = df_word.query('count >=2')
df_word
```

```python
# 빈도표 만들기
df_word = df_word.groupby('word', as_index = False) \
                .agg( n = ('word','count')) \
                .sort_values('n', ascending=False)

df_word
```

```python
# 단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

```python
# 가로 세로 크기 설정
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update( { 'figure.figsize' : [6.5, 6],
                     'font.family' : 'Malgun Gothic'})

#막대 그래프 만들기
sns.barplot( data= top20, x='n', y='word');
```

```python
# wc 만들기

# 1. 데이터 프레임을 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
# 워드클라우드 만들기
import PIL
icon = PIL.Image.open('cloud.png')

font = 'DoHyeon-Regular.ttf'

import numpy as np
img = PIL.Image.new('RGB', icon.size,(255,255,255))
img.paste(icon, icon)
img = np.array(img)
```

```python
# 2. wc 만들기
from wordcloud import WordCloud
wc = WordCloud( random_state = 1234,
              font_path = font,
              width = 400,
              height = 400,
              background_color = 'white',
              mask = img)
```

```python
# 워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.figure(figsize = (10,10)) #가로, 세로 크기 설정
plt.axis('off')  #테두리 선 없애기
plt.imshow(img_wordcloud);  #워드 클라우드 출력
```
