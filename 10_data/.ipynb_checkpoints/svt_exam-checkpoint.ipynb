```python
# 가사 불러오기
svt = open('svt_pretty.txt', encoding = 'UTF-8').read()
svt
```

```python
# 불필요한 문자 제거
import re
svt = re.sub('[^가-힣]', ' ', svt)
svt
```

```python
# 명사 추출
# 1. 형태소 만들기

import konlpy
hannanum = konlpy.tag.Hannanum()

# 2. 명사 추출
hannanum.nouns("세븐틴 그룹의 멤버는 총 13명이다.")
```

```python
# 노래 가사에서 명사 추출
nouns = hannanum.nouns(svt)
nouns
```

```python
# 데이터 프레임으로 변환
import pandas as pd
df_word = pd.DataFrame( { 'word' : nouns})
df_word
```

```python
# 단어 빈도표 만들기
# 글자 수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
# 두 글자 이상 단어만 남기기
df_word = df_word.query('count >=2')
df_word.sort_values('count')
```

```python
# 단어 빈도 구하기
df_word = df_word.groupby('word', as_index = False) \
                .agg(n = ('word','count')) \
                .sort_values('n', ascending = False)
df_word
```

```python
# 상위 20개
top20 = df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update( { 'font.family' : 'Malgun Gothic' ,
                     'figure.dpi' : '120',
                     'figure.figsize' : [6.5, 6]})

# 막대 그래프 만들기
sns.barplot( data = top20, x = 'n', y='word')
```

```python
font = 'DoHyeon-Regular.ttf'
```

```python
# 데이터 프레임을 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
# 워드클라우드 만들기
import PIL
icon = PIL.Image.open('cloud.png')
```

```python
import numpy as np
img = PIL.Image.new('RGB', icon.size,(255,255,255))
img.paste(icon, icon)
img = np.array(img)
```

```python
from wordcloud import WordCloud
wc = WordCloud( random_state = 1234,  #난수 고정
              font_path = font,       #폰트 설정
              width = 400,            #가로 크기
              height = 400,           #세로 크기
              background_color = 'white',  #배경색
              mask = img)             #mask 설정
```

```python
# 워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.figure(figsize = (10,10)) #가로, 세로 크기 설정
plt.axis('off')  #테두리 선 없애기
plt.imshow(img_wordcloud);  #워드 클라우드 출력
```
