```python
moon = open('speech_moon.txt', encoding = 'UTF-8').read()
moon
```

```python
# 불필요한 문자 제거하기
import re
moon = re.sub('[^가-힣]', ' ', moon)
```

```python
#hananaum 만들기
import konlpy
hannanum = konlpy.tag.Hannanum()

# 명사 추출하기
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
# 연설문에서 명사 추출하기
nouns = hannanum.nouns(moon)
nouns
```

```python
# 데이터 프레임으로 변환
import pandas as pd
df_word = pd.DataFrame( { 'word' : nouns})
df_word
```

```python
# 글자 수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
# 두 글자 이상 단어만 남겨두기
df_word = df_word.query( 'count >= 2')
df_word.sort_values('count')
```

```python
# 단어 빈도 구하기
df_word = df_word.groupby('word', as_index = False ) \
                .agg(n=('word','count')) \
                .sort_values('n', ascending = False)

df_word
```

```python
# 단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.rcParams.update( { 'font.family' : 'Malgun Gothic',
                    'figure.dpi' : 120 ,
                    'figure.figsize' : [6.5 , 6]})
                

# 막대 그래프 만들기
sns.barplot( data = top20, y='word',x='n');
        
```

```python
font = 'DoHyeon-Regular.ttf'
```

```python
#데이터 프레임을 딕셔너리로 변환
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

```python
# wc만들기
from wordcloud import WordCloud
wc = WordCloud( random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white')
```

```python
# 워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.figure(figsize = (10,10)) #가로 세로 크기 설정
plt.axis('off')  #테두리 선 없애기
plt.imshow(img_wordcloud); #워드클라우드 출력
```

```python
import PIL
icon = PIL.Image.open('cloud.png')
```

```python
import numpy as np
img = PIL.Image.new('RGB',icon.size, (255,255,255))
img.paste(icon, icon)
img = np.array(img)
```

```python
# wc만들기
from wordcloud import WordCloud
wc = WordCloud( random_state = 1234,
               font_path = font,
               width = 400,
               height = 400,
               background_color = 'white',
              mask = img)
```

```python
# 워드 클라우드 만들기
img_wordcloud = wc.generate_from_frequencies(dic_word)

# 워드 클라우드 출력하기
plt.figure(figsize = (10,10)) #가로 세로 크기 설정
plt.axis('off')  #테두리 선 없애기
plt.imshow(img_wordcloud); #워드클라우드 출력
```
