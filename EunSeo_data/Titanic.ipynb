```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# 생존자/ 사망자 구하기
df['Survived'].value_counts()
```

```python
# 생존여부 칼럼 추가
df['Survived2'] = df['Survived'].replace(0, 'dead').replace(1, 'survivor')

df
```

```python
sns.countplot(data=df, x='Survived2', hue='Sex')
```

```python
# 클래스 칼럼 추가

df['Pclass2'] = df['Pclass'].replace(1, 'First Class').replace(2, 'Business').replace(3, 'Economy')
df
```

```python
sns.countplot(data=df, x='Pclass2', hue='Survived2');
```

```python
# 생애주기별 기준으로 나이 나누기

df['Age2'] = np.where( df['Age'] >= 65 , 'old',
             np.where(df['Age'] >= 30, 'middle',
             np.where(df['Age']>= 20 , 'youth',
                      np.where(df['Age']>=0, 'young', 'Nan'))))
df
```

```python
sns.countplot(data=df, x='Age2');
```
